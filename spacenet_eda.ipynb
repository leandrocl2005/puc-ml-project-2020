<h2>Exploração dos dados spacenet</h2>

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import glob
from skimage import io
%matplotlib inline 
```

```python
png_files = glob.glob('spacenet/PNG-RGB/*.png')
tif_files = glob.glob('spacenet/TIF-RGB/*.tif')
mask_png_files = glob.glob('spacenet/PNG-MASK/*.png')
```

```python
for i, file in enumerate(tif_files):
    plt.subplot(3, 5, i + 1)
    img = io.imread(file)
    plt.imshow(img)
    plt.axis('off')
for i, file in enumerate(png_files):
    plt.subplot(3, 5, i + 6)
    img = io.imread(file)
    plt.imshow(img)
    plt.axis('off')
for i, file in enumerate(mask_png_files):
    plt.subplot(3, 5, i + 11)
    img = io.imread(file)
    plt.imshow(img)
    plt.axis('off')
plt.show()
```

```python
png_file = png_files[1]
img = Image.open(png_file).convert("L")
img_array = np.array(img) 
img.close()
    
mask_file = mask_png_files[1]
img = Image.open(mask_file)
mask_array = np.array(img) 
img.close()

fig = plt.figure(1, figsize = [20, 10], dpi = 200)
plt.subplot(1,2,1)
plt.imshow(mask_array)
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(img_array)
plt.axis("off")
plt.show()
```
